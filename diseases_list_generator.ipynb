```python
import pandas as pd
import numpy as np
```

```python
root = '../'
```

```python
import glob
```

```python
path = root + 'CSV/TabNet/Internacoes_Rate/'
all_files = glob.glob(path + "*.csv")
csv_files = []
for file in all_files:
    csv_files.append(file.split("\\")[1].split(".csv")[0])
csv_files
```

```python
mapper = pd.DataFrame()
mapper['CSV'] = csv_files
```

```python
def toCamelCase(field):
    return field.replace('_', ' ').title()

mapper['SELECT_NAME'] = mapper['CSV'].apply(toCamelCase)
```
